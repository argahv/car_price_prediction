```python
from sklearn.impute import SimpleImputer


mean_imputer = SimpleImputer(missing_values=np.nan,strategy="mean")
median_imputer = SimpleImputer(missing_values=np.nan,strategy="median")

```

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python

dataset.isnull().sum()
```

```python
dataset_to_clean = dataset.copy()
dataset_to_clean
```

## Engine Fuel Type 

```python
print(dataset_to_clean["Engine Fuel Type"].value_counts())
print(sum(dataset_to_clean['Engine Fuel Type'].value_counts()))
```

```python
print(dataset_to_clean["Engine Fuel Type"].unique())
# Mapping unique fuel type based on index

engine_fuel_type_mapping = {label:index for index,label in enumerate(dataset_to_clean['Engine Fuel Type'].unique())}
print(engine_fuel_type_mapping)
```

```python
dataset_to_clean["Engine Fuel Type"] = dataset_to_clean["Engine Fuel Type"].map(engine_fuel_type_mapping)
```

```python
index_eft = np.where(dataset_to_clean.loc[:,"Engine Fuel Type"]==10)
print(index_eft)
dataset_to_clean.loc[index_eft[0][0]:(index_eft[0][0]+10),:]

dataset_to_clean.loc[index_eft[0],'Engine Fuel Type'] = 1
```

```python
# Check the category which corresponds to NA has been removed
print(dataset_to_clean['Engine Fuel Type'].unique())
```

## Engine HP

```python
sns.boxplot( y="Engine HP", data=dataset)
```

```python
mean_imputer.fit(dataset_to_clean[["Engine HP"]])
dataset_to_clean["Engine HP"] =  mean_imputer.transform(dataset_to_clean[["Engine HP"]])
print(print(dataset_to_clean["Engine HP"].isnull().sum()))
```

## Engine Cylinder

```python
print(dataset_to_clean['Engine Cylinders'].value_counts(), "\n")
print(sum(dataset_to_clean['Engine Cylinders'].value_counts()))
```

```python
index_eng_cyl = np.where(dataset_to_clean["Engine Cylinders"].isnull())
print('Missing values',len(index_eng_cyl[0]))
```

```python
missing_eng_cars = pd.DataFrame(dataset_to_clean.loc[index_eng_cyl[0],["Make","Model"]].values,columns=["Make","Model"])
print(missing_eng_cars)
```

```python
# Solution to reduce redundancy
list(missing_eng_cars.groupby(['Make','Model']).groups.keys())
```

```python
#all of the cars do not have cylinders so
dataset_to_clean.loc[index_eng_cyl[0][0:10],'Engine Cylinders'] = 0 
dataset_to_clean.loc[index_eng_cyl[0][10:],'Engine Cylinders'] = 0
```

## Number of doors

```python
indexes_nb_doors = np.where(dataset_to_clean['Number of Doors'].isnull())
dataset_to_clean.loc[indexes_nb_doors]
```

```python
# Ferrari and teslas model s has 3 and 5 doors so
dataset_to_clean.loc[indexes_nb_doors[0][0],'Number of Doors'] = 3
dataset_to_clean.loc[indexes_nb_doors[0][1:],'Number of Doors'] = 5
```

## MArket category

```python
dataset_to_clean.loc[np.where(dataset_to_clean['Market Category'].isnull())[0],'Market Category'] = 'No category'

# 5 most frequent categories
dataset_to_clean['Market Category'].value_counts().head(5)
```
